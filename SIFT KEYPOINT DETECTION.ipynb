```python
# I/P img original form nd also i/p img converted into grayscale for SIFT
import cv2
import matplotlib.pyplot as plt

# Step 1: Load the original images
img1 = cv2.imread('aero1.jpg')
img2 = cv2.imread('aero2.jpg')

# Step 2: Convert to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Step 3: Convert originals to RGB for proper color display
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Step 4: Display Original + Grayscale Images
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img1_rgb)
plt.title('Image 1 (Original)')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(img2_rgb)
plt.title('Image 2 (Original)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(gray1, cmap='gray')
plt.title('Image 1 (Grayscale)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(gray2, cmap='gray')
plt.title('Image 2 (Grayscale)')
plt.axis('off')


plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Load two grayscale images
img1 = cv2.imread('aero1.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('aero2.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Create SIFT detector
sift = cv2.SIFT_create()

# Step 3: Detect keypoints and descriptors
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)

# Step 4: Brute-Force matcher : for detecting To match keypoints between two images.
bf = cv2.BFMatcher()
matches = bf.match(des1, des2)

# Step 5: Sort matches based on distance
matches = sorted(matches, key=lambda x: x.distance)

# Step 6: Draw first 50 matches
matched_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], None)

# Step 7: Display the result
plt.figure(figsize=(14, 7))
plt.imshow(matched_img, cmap='gray')
plt.title('SIFT Keypoint Matching Between Two Images')
plt.axis('off')
plt.show()
```
