```python
#Implement the Harris corner detector to find and visualize corners in a grayscale image.


import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load and convert to grayscale
img = cv2.imread('flower.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

# Harris detector
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
dst = cv2.dilate(dst, None)

# Threshold and draw circles for corner points
threshold = 0.01 * dst.max()
corner_img = img.copy()

# Draw circles at corner points
for y in range(dst.shape[0]):
    for x in range(dst.shape[1]):
        if dst[y, x] > threshold:
            cv2.circle(corner_img, (x, y), 3, (0, 0, 255), -1)  # radius = 3

# Display
plt.imshow(cv2.cvtColor(corner_img, cv2.COLOR_BGR2RGB))
plt.title('Harris Corners with Bigger Dots')
plt.axis('off')
plt.show()
```
